```python
#Import some libraries 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Compute the numebr of significant changes in the perturbation matrix for in each scenario you want
```

```python
#function to compute the number of values in the matrix that denote a significant change in firing rates
def count_entry(matrix):
    count=0
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if matrix[i][j] != 0 :
                count+=1
    print(count)
    return count
```

```python
#function to compute the number of values in the matrix that denote a positive change in firing rates
def count_positive(matrix):
    count=0
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if matrix[i][j] ==1  :
                count+=1
    print(count)
    return count

#function to compute the number of values in the matrix that denote a negative change in firing rates
def count_negative(matrix):
    count=0
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if matrix[i][j] == -1  :
                count+=1
    print(count)
    return count
```

```python
#Number of  counts for each matrix (I call the function I just created)
#I am uploading all the matrices (spontanous, FF,FB, FF+FB. different FB conditions)
count_spont=count_entry(np.loadtxt('../simulations_data/matrix/0.2_sign_In_spontaneuous.txt'))
count_ff_e4= count_entry(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_e4.txt'))
count_e4_e5= count_entry(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_e4e5.txt'))

#different FB conditions
count_fb_e5= count_entry(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_e5.txt'))
count_fb_e23_e5= count_entry(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_Feedback_In_with_e23_e5_half_each.txt'))
count_fb_e6= count_entry(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_Feedback_In_with_e6.txt'))
count_fb_e5_pv5= count_entry(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_Feedback_In_with_e5_pv5_half_each.txt'))
count_fb_e23= count_entry(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_Feedback_In_with_e23.txt'))

#I save the values in an list
counts_all=[count_spont,count_ff_e4,count_e4_e5,count_fb_e5,count_fb_e23,count_fb_e23_e5,count_fb_e5_pv5,count_fb_e6]
```

```python
#Figure for all confiugurations with counts of significant changes
fig , ax = plt.subplots(figsize=(14,6))
labels_all = ('Spont','FF e4','FF e4+FB e5','FB e5','FB e2/3','FB e2/3_e5','FB e5_pv5','FB e6')

#labels on x axis
x_pos = np.arange(len(labels_all))
plt.xticks(x_pos, labels_all,size=15)

#plot data 
plt.bar(x_pos,counts_all)

#layout
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.setp(ax.spines.values(), linewidth=3)
plt.tick_params(axis='y', labelsize=15)
ax.xaxis.set_tick_params(width=3)
ax.yaxis.set_tick_params(width=3)

#labels of the axes
plt.ylabel('# significant changes', size=25) #The possible total for each is 256

plt.show()
```

```python
#Number of poisitve counts for each matrix (I call the function of positive changes count)
#I am uploading all the matrices and computing it
p_count_spont=count_positive(np.loadtxt('../simulations_data/matrix/0.2_sign_In_spontaneuous.txt'))
p_count_ff_e4= count_positive(np.loadtxt('../simulations_data/matrix/sign_inpute4_20'))
p_count_e4_e5= count_positive(np.loadtxt('../simulations_data/matrix/sign_inpute4_e5_20perc'))

p_count_fb_e5= count_positive(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_e5.txt'))
p_count_fb_e23_e5= count_positive(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_Feedback_In_with_e23_e5_half_each.txt'))
p_count_fb_e6= count_positive(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_Feedback_In_with_e6.txt'))
p_count_fb_e5_pv5= count_positive(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_Feedback_In_with_e5_pv5_half_each.txt'))
p_count_fb_e23= count_positive(np.loadtxt('../simulations_data/matrix/0.2_sign_In_with_Feedback_In_with_e23.txt'))

#I create an array with these values 
p_counts_all=[p_count_spont,p_count_ff_e4,p_count_e4_e5,p_count_fb_e5,p_count_fb_e23,p_count_fb_e23_e5,p_count_fb_e5_pv5,p_count_fb_e6]
print(p_counts_all)

#I compute the percentage of positive counts over the total counts of significant changes
counts_perc_all= [(a / b)*100 for a, b in zip(p_counts_all, counts_all)]
print(counts_perc_all)
```
