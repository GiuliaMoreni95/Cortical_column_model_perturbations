```python
#Import some libraries 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Compute the numebr of significant changes in the perturbation matrix for in each scenario you want
```

```python
#function to compute the number of values in the matrix that denote a significant change in firing rates
def count_entry(matrix):
    count=0
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if matrix[i][j] != 0 :
                count+=1
    print(count)
    return count
```

```python
#function to compute the number of values in the matrix that denote a positive change in firing rates
def count_positive(matrix):
    count=0
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if matrix[i][j] ==1  :
                count+=1
    print(count)
    return count

#function to compute the number of values in the matrix that denote a negative change in firing rates
def count_negative(matrix):
    count=0
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if matrix[i][j] == -1  :
                count+=1
    print(count)
    return count
```

```python
#Number of  counts for each matrix (I call the function I just created)
#I am uploading all the matrices (spontanous, FF,FB, FF+FB. different FB conditions)
count_spont=count_entry(np.loadtxt('../simulations_data/REVISIONS_1/matrix/sign_baseline_20%.txt'))
count_ff_e4= count_entry(np.loadtxt('../simulations_data/REVISIONS_1/matrix/sign_L4_20%.txt'))
count_e4_e5= count_entry(np.loadtxt('../simulations_data/REVISIONS_1/matrix/sign_L4_L5_20%.txt'))

#different FB conditions
count_fb_e5= count_entry(np.loadtxt('../simulations_data/REVISIONS_1/matrix/0.2_sign_Sim_e4_0pA_e5_150pA_pert_30pA.txt'))
count_fb_e23_e5= count_entry(np.loadtxt('../simulations_data/REVISIONS_1/matrix/sign_L5_L23_20%.txt'))
count_fb_e6= count_entry(np.loadtxt('../simulations_data/REVISIONS_1/matrix/sign_L6_20%.txt'))
count_fb_e23= count_entry(np.loadtxt('../simulations_data/REVISIONS_1/matrix/sign_L23_20%.txt'))

count_fb_e23_e5_e6= count_entry(np.loadtxt('../simulations_data/REVISIONS_1/matrix/sign_L5_L23_L6_20%.txt'))



#I save the values in an list
counts_all=[count_spont,count_ff_e4,count_fb_e5,count_e4_e5,count_fb_e23,count_fb_e23_e5,count_fb_e23_e5_e6,count_fb_e6]
```

```python
#Figure for all confiugurations with counts of significant changes
fig , ax = plt.subplots(figsize=(14,6))
labels_all = ('Spont','e4','e5','e4,e5','e2/3','e2/3,e5','e2/3,e5,e6','e6')

#labels on x axis
x_pos = np.arange(len(labels_all))
plt.xticks(x_pos, labels_all,size=15)

#plot data 
plt.bar(x_pos,counts_all)

#layout
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.setp(ax.spines.values(), linewidth=3)
plt.tick_params(axis='y', labelsize=15)
ax.xaxis.set_tick_params(width=3)
ax.yaxis.set_tick_params(width=3)

#labels of the axes
plt.ylabel('# changes > 20%', size=25) #The possible total for each is 256

plt.show()
```
