```python
def organize_spike_times(N, spike_times, neuron_indices):
    """
    Organize spike times by neuron.

    Parameters:
    - N (int): Number of neurons.
    - spike_times (list or array): Times at which spikes occurred.
    - neuron_indices (list or array): Indices of neurons that spiked at the corresponding times in spike_times.

    Returns:
    - list of lists: A list where each sublist contains the spike times for a corresponding neuron.
    """
    # Initialize a list of empty lists for each neuron
    neuron_spike_times = [[] for _ in range(N)]

    # Populate the list with spike times for each neuron
    for time, neuron_index in zip(spike_times, neuron_indices):
        #print(neuron_index)
        neuron_index = int(neuron_index)  # Ensure neuron index is an integer

        neuron_spike_times[neuron_index].append(time)

    return neuron_spike_times

def compute_ISI(N, spike_times, neuron_indices):
    """
    Compute the inter-spike intervals (ISI) for a given set of neurons.

    Parameters:
    - N (int): Number of neurons.
    - spike_times (list or array): Times at which spikes occurred.
    - neuron_indices (list or array): Indices of neurons that spiked at the corresponding times in spike_times.

    Returns:
    - list of lists: A list where each element is a list of ISIs for a neuron.
    """
    # Organize spike times by neuron
    neuron_spike_times = organize_spike_times(N, spike_times, neuron_indices)

    # Calculate ISI for each neuron
    ISI = [[] for _ in range(N)]
    for n in range(N):
        ISI[n] = [neuron_spike_times[n][i + 1] - neuron_spike_times[n][i] 
                  for i in range(len(neuron_spike_times[n]) - 1)]

    return ISI

# Example usage
spike_times = [0.1, 0.2, 0.5, 0.6, 1.0, 1.2, 1.5, 1.8, 2.0]  # Example spike times
neuron_indices = [0, 1, 0, 1, 0, 1, 0, 1, 0]                # Corresponding neuron indices

# Compute ISI
isi = compute_ISI(2, spike_times, neuron_indices)
print(isi)
```

```python
def compute_CV(ISI):
    """
    Compute the coefficient of variation (CV) for inter-spike intervals (ISI).
    
    Parameters:
    - ISI (list of lists): A list where each element is a list of ISIs for a neuron.
    
    Returns:
    - list: A list where each element is the CV for a neuron.
    """
    ISI = [x for x in ISI if x != []]  # Remove empty ISI lists (neurons without spikes)
    cvs = [np.std(i) / np.mean(i) if np.mean(i) > 0 else 0 for i in ISI]  # Compute CV for each neuron
    return cvs
```

```python
def save_CV_to_file(cvs, filename):
    """
    Save the CVs to a text file.
    
    Parameters:
    - cvs (list): List of CV values.
    - filename (str): Path to the output text file.
    """
    np.savetxt(filename, cvs)

```
